```python
from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("alvaroalon2/biobert_genetic_ner")

model = AutoModelForTokenClassification.from_pretrained("alvaroalon2/biobert_genetic_ner")
```

```python
from transformers import pipeline
nlp = pipeline("ner", model=model, tokenizer=tokenizer,grouped_entities=True)
  
  
  
```

```python
with open('/content/eye color.txt', 'r') as file:
    data = file.read()
```

```python
chunks=data.split("\n")
ner_results=[]
import re
```

```python
for chunk in chunks:
  
  #chunk.replace("\n"," ")
  chunk = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", chunk )
  ner_result=nlp(chunk)
  if(ner_result!=[]):
     ner_results.append(ner_result)
```

```python
flat_list= []
for sublist in ner_results:
    for item in sublist:
        flat_list.append(item)
```

```python
import pandas as pd
df1 = pd.DataFrame(flat_list)
df1
```

```python
# pip install -U spacy
# python -m spacy download en_core_web_sm
import spacy

# Load English tokenizer, tagger, parser and NER
nlp_e = spacy.load("en_core_web_sm")

# Process whole documents
doc = nlp_e(data)
eth=[]
# Find named entities, phrases and concepts
for entity in doc.ents:
    if entity.label_=="NORP":
      if entity.text not in eth:
         eth.append(entity.text)
         print(entity.text)
```

```python
import pandas as pd
df = pd.DataFrame(eth)
df
```

```python
result = pd.concat([df1, df], axis=1)
```
